```python
import json
import os

# Ruta al archivo json
file_path = os.path.abspath(os.path.join(os.getcwd(), '../../demo_placeholder.json'))

# Cargar los datos
with open(file_path, 'r', encoding='utf-8') as f:
    demo_data = json.load(f)

# Eliminar _id si existe
if '_id' in demo_data:
    del demo_data['_id']

# Actualizar el documento con slug 'demo'
result = platforms.replace_one({'slug': 'demo'}, demo_data, upsert=True)

print(f"Demo actualizada. Matched: {result.matched_count}, Modified: {result.modified_count}")



```
