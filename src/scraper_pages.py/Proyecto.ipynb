```python
#Import lectura HTML
from utils.requestHTTP import fetch_html

#Import manejo de urls desde el HTML
from utils.urls_extractor_from_html import urls_extractor_from_html  
from utils.urls_utilities_cleaner import urls_utilities_cleaner
from utils.urls_format_from_domain import urls_format_from_domain

#import manejo de html
from utils.html_spliter_head_header_main_footer import html_spliter_head_header_main_footer

#import manejo de textos desde el html
from utils.text_extractor_from_html import text_extractor_from_html
```

```python
#Lectura del HTML
link ="https://reental.co"
html = await fetch_html(link)
#print(html)
```

```python
head_main_foot = html_spliter_head_header_main_footer(html)
```

```python
head = head_main_foot["head"]
#print(head[:1000])
head_text = text_extractor_from_html(head)
head_urls = urls_extractor_from_html(head)
head_urls = urls_utilities_cleaner(head_urls)
head_urls = urls_format_from_domain(head_urls, link)

print("TEXTOS")
print("Cantidad de textos: " + str(len(head_text)))
print(head_text[:20])
print("URLs")
print("Cantidad de urls: " + str(len(head_urls)))
print(head_urls[:20])
```

```python
header = head_main_foot["header"]
#print(header[:1000])
header_text = text_extractor_from_html(header)
header_urls = urls_extractor_from_html(header)
header_urls = urls_utilities_cleaner(header_urls)
header_urls = urls_format_from_domain(header_urls, link)

print("TEXTOS")
print("Cantidad de textos: " + str(len(header_text)))
print(header_text[:20])
print("URLs")
print("Cantidad de urls: " + str(len(header_urls)))
print(header_urls[:20])
```
